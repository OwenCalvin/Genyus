# Ouverture du fichier contenant toutes les musiques

```python
from typing import List
import numpy as np
import json

songs: List[dict] = []
french: List[str] = []

with open("../../src/datas/songs.json", "r", encoding="utf8") as file:
        songs = json.loads(file.read())

with open("../../src/datas/french.txt", "r", encoding="utf8") as file:
        french = file.read().lower().splitlines()
```

```python
unknown_words: dict = {}

for song in songs:
    lyrics = song["lyrics_dictionary"]
    words: List[str] = lyrics.split(" ")

    for word in words:
        if word not in french:
            if word in unknown_words:
                unknown_words[word] += 1
            else:
                unknown_words[word] = 0
    
print(unknown_words)
```
