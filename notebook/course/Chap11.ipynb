<h1 style="color:red;padding:4px;border-bottom: 4px solid red">INTSLIDER</h1>

**Exemple :**
```python
# 0) Import required packages
import ipywidgets as ipw
import numpy as np
import matplotlib.pyplot as plt

# 1) Create the widget
nb_points_slider = ipw.IntSlider(value=500, min=0, max=1000, step=10, description='Nb points:')

# 2) Function called when widget's value is changed
def nb_points_changed(nb_points):
    plt.style.use("seaborn-talk")
    plt.figure(figsize=(20,20))
    plt.title("Graph with Points")
    plt.xlabel("X")
    plt.ylabel("Y")

    X = np.random.randn(nb_points)
    Y = np.random.randn(nb_points)*100

    # NOTE: we use random.random here to make sure values are > 0 and we rescale
    S = np.random.random(nb_points)*250

    # We use a colormap and we pass an array-like variable storing indexes to this colormap
    colormap = plt.get_cmap("plasma")
    colors = np.random.random(nb_points)*500

    plt.scatter(X, Y, s=S, c=colors, cmap=colormap)
    plt.colorbar()

# 3) Connect widget and function AND display GUI
ipw.interact(nb_points_changed, nb_points = nb_points_slider);
```

<h1 style="color:red;padding:4px;border-bottom: 4px solid red">CONTRÔLER L'AFFICHAGE</h1>

## la fonction **[`interactive_output`](https://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html#More-control-over-the-user-interface:-interactive_output)**
### Séparer l'étape de création de l'interaction et l'affichage
### Passer les interactions entre widgets et paramètres de fonctions sous forme de **`dict`**.

**NOTE :** il existe beaucoup de fonctionalités pour **[changer l'aspect de la GUI](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20Styling.html)**
. Ce n'est pas l'objectif de ce cours.

**Exemple :**
```python
# 0) Import required packages
import ipywidgets as ipw
import numpy as np
import matplotlib.pyplot as plt

# 1) Create the widgets
nb_points_slider = ipw.IntSlider(value=500, min=0, max=1000, step=10, description='Nb points:')
nb_points_text = ipw.IntText(value=500)

# NOTE: nb_points_text's display can be ignored here now
# display(nb_points_text)

# Bi-directional linking
ipw.link((nb_points_slider, 'value'), (nb_points_text, 'value'))
ipw.link((nb_points_text, 'value'), (nb_points_slider, 'value'))

colorbar_toggle = ipw.ToggleButton(value=True, description='Colorbar', button_style='', tooltip='Display colorbar (or not)', icon='eye')
colormap_dropdown = ipw.Dropdown(options=['plasma', 'Reds', 'Greens', 'Blues'], value='plasma', description='Colormap:', disabled=False)

# Set the layout
ui = ipw.HBox([nb_points_slider, nb_points_text, colorbar_toggle, colormap_dropdown])

# 2) Function called when widget's value is changed
def scatter_graph_changed(nb_points, display_color_bar, colormap):
    plt.style.use("seaborn-talk")
    plt.figure(figsize=(20,20))
    plt.title("Graph with Points")
    plt.xlabel("X")
    plt.ylabel("Y")

    X = np.random.randn(nb_points)
    Y = np.random.randn(nb_points)*100

    # NOTE: we use random.random here to make sure values are > 0 and we rescale
    S = np.random.random(nb_points)*250

    # We use a colormap and we pass an array-like variable storing indexes to this colormap
    #colormap = plt.get_cmap("plasma")
    colors = np.random.random(nb_points)*500

    plt.scatter(X, Y, s=S, c=colors, cmap=colormap)
    if display_color_bar:
        plt.colorbar()

# 3) Connection widget --> function
interactions = ipw.interactive_output(scatter_graph_changed, {'nb_points':nb_points_slider, 'display_color_bar':colorbar_toggle, 'colormap':colormap_dropdown});

# 4) Now display ALL
display(ui, interactions)
```
